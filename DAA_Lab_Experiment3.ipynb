```python
import heapq


# ==========================================
# UNION-FIND FOR KRUSKAL
# ==========================================

class UnionFind:

    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n


    def find(self, x):

        if self.parent[x] != x:
            # Path compression
            self.parent[x] = self.find(self.parent[x])

        return self.parent[x]


    def union(self, x, y):

        rx = self.find(x)
        ry = self.find(y)

        if rx == ry:
            return False

        # Union by rank
        if self.rank[rx] < self.rank[ry]:
            rx, ry = ry, rx

        self.parent[ry] = rx

        if self.rank[rx] == self.rank[ry]:
            self.rank[rx] += 1

        return True


# ==========================================
# KRUSKAL'S ALGORITHM
# ==========================================

def kruskal(n, edges):
    """
    edges: list of (weight, u, v)
    """

    # Sort edges by weight
    # O(E log E)
    edges.sort()

    uf = UnionFind(n)

    mst = []
    cost = 0

    for w, u, v in edges:

        # Add edge only if it doesn't create a cycle
        if uf.union(u, v):

            mst.append((u, v, w))
            cost += w

            # MST contains n - 1 edges
            if len(mst) == n - 1:
                break

    return mst, cost


# ==========================================
# PRIM'S ALGORITHM
# ==========================================

def prim(n, adj, start=0):
    """
    adj: adjacency list {u: [(v, w), ...]}
    """

    INF = float('inf')

    # Minimum edge weight to connect each vertex
    key = [INF] * n

    # Stores parent of each vertex
    parent = [-1] * n

    # Tracks vertices already included in MST
    inMST = [False] * n

    # Start vertex
    key[start] = 0

    # Priority queue
    pq = [(0, start)]

    mst = []
    cost = 0


    while pq:

        # Get vertex with minimum weight
        w, u = heapq.heappop(pq)

        # Skip if already included
        if inMST[u]:
            continue

        # Add vertex to MST
        inMST[u] = True


        # Add edge to MST
        if parent[u] != -1:

            mst.append((parent[u], u, w))

            cost += w


        # Check all adjacent vertices
        for v, wt in adj.get(u, []):

            if not inMST[v] and wt < key[v]:

                key[v] = wt

                parent[v] = u

                heapq.heappush(
                    pq,
                    (wt, v)
                )


    return mst, cost


# ==========================================
# GRAPH DEFINITION
# ==========================================

n = 7

edges = [
    (7, 0, 1),
    (5, 0, 3),
    (8, 1, 2),
    (9, 1, 3),
    (7, 1, 4),
    (5, 2, 4),
    (15, 3, 4),
    (6, 3, 5),
    (8, 4, 5),
    (9, 4, 6),
    (11, 5, 6)
]


# ==========================================
# CREATE ADJACENCY LIST
# ==========================================

adj = {}

for w, u, v in edges:

    adj.setdefault(u, []).append((v, w))

    adj.setdefault(v, []).append((u, w))


# ==========================================
# RUN KRUSKAL
# ==========================================

k_mst, k_cost = kruskal(
    n,
    edges[:]
)


# ==========================================
# RUN PRIM
# ==========================================

p_mst, p_cost = prim(
    n,
    adj
)


# ==========================================
# PRINT KRUSKAL RESULT
# ==========================================

print("=== Kruskal's MST ===")

for u, v, w in k_mst:

    print(
        f"Edge ({u} - {v}) Weight: {w}"
    )

print(
    f"Total MST Cost: {k_cost}"
)


# ==========================================
# PRINT PRIM RESULT
# ==========================================

print("\n=== Prim's MST ===")

for u, v, w in p_mst:

    print(
        f"Edge ({u} - {v}) Weight: {w}"
    )

print(
    f"Total MST Cost: {p_cost}"
)
```
