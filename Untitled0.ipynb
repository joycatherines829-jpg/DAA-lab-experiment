```python
import heapq

def dijkstra(graph, source):
    """
    Dijkstra's Algorithm using Min-Heap
    Time: O((V + E) log V)
    Space: O(V)

    graph: dict {u: [(v, weight), ...]}
    """

    n = len(graph)

    # Distance from source to every vertex
    dist = [float('inf')] * n

    # Previous vertex used to reconstruct shortest path
    prev = [None] * n

    # Distance from source to itself is 0
    dist[source] = 0

    # Min-heap: (distance, vertex)
    pq = [(0, source)]

    # Set of visited vertices
    visited = set()

    while pq:
        # Get vertex with smallest distance
        d, u = heapq.heappop(pq)

        # Skip if already visited
        if u in visited:
            continue

        visited.add(u)

        # Check all neighbors of u
        for v, w in graph[u]:

            # Relaxation
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                prev[v] = u

                # Add updated distance to heap
                heapq.heappush(pq, (dist[v], v))

    return dist, prev


def reconstruct_path(prev, source, target):
    path = []

    node = target

    # Move backwards from target to source
    while node is not None:
        path.append(node)
        node = prev[node]

    # Reverse to get source -> target
    path.reverse()

    # Check if path actually starts from source
    if path[0] == source:
        return path

    return []


# --- Graph Definition (Adjacency List) ---

graph = {
    0: [(1, 4), (2, 1)],
    1: [(3, 1)],
    2: [(1, 2), (3, 5)],
    3: [(4, 3)],
    4: [(5, 2)],
    5: []
}

# Starting vertex
source = 0

# Run Dijkstra's algorithm
dist, prev = dijkstra(graph, source)

# Print results
print(f"Shortest paths from vertex {source}:")

print(f'{"Vertex":>8} {"Distance":>10} {"Path":>30}')

print('-' * 55)

for v in range(len(graph)):

    # Get shortest path
    path = reconstruct_path(prev, source, v)

    # Convert path list to string
    path_str = ' -> '.join(map(str, path)) if path else 'No path'

    # Get distance
    d = dist[v] if dist[v] != float('inf') else 'INF'

    # Print result
    print(f'{v:>8} {str(d):>10} {path_str:>30}')
```
